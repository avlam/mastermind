```python
pins_df=pd.read_csv("four-digit-pin-codes-and-frequency.csv")
#print(pins_df)
pins_df.shape
pins_df=pins_df['Pins'].str.split(';', 1, expand=True)
#pins_df
```

```python
# Rename columns and group by frequency
new_pins_df=pins_df.rename(columns={0: "Pins", 1: "Frequency"})
# new_pins_df
New_pins_df=new_pins_df.apply(pd.to_numeric)
# New_pins_df.dtypes
#New_pins_df
```
